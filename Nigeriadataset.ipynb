```python
url0 = "https://datasets.mojaglobal.workers.dev/0:/Administrative/Boundaries/Level4%20by%20State/NGA/NGA_ALL_states.json"

a=requests.get(url0).content

df_cbo = gpd.read_file(io.StringIO(a.decode('utf-8')))
df_cbo
```

```python
url0 = "https://datasets.mojaglobal.workers.dev/0:/Administrative/Boundaries/Level4%20by%20State/NGA/NGA_ALL_states.json"

a=requests.get(url0).content

df1 = gpd.read_file(io.StringIO(a.decode('utf-8')))
fig = plt.figure(1, figsize=(25,15)) 
ax = fig.add_subplot()
df1.apply(lambda x: ax.annotate(s=x.locname, xy=x.geometry.centroid.coords[0], ha='center', fontsize=14),axis=1);
df1.boundary.plot(ax=ax, color='Black', linewidth=.4)
df1.plot(ax=ax, cmap='Pastel2', figsize=(12, 12))
ax.text(-0.05, 0.5, 'Map of Nigeria showing its states', transform=ax.transAxes,
        fontsize=20, color='gray', alpha=0.5,
        ha='center', va='center', rotation='90')
```

```python
url2 = "https://datasets.mojaglobal.workers.dev/0:/Soil/World%20Soil%20Resources/byStateWSR/NGA/NGA_ALL_states_WSR.json"

s=requests.get(url2).content

df2 = gpd.read_file(io.StringIO(s.decode('utf-8')))
df2
```

```python

ax = sns.displot(df2, x='name', kind='hist', hue='IPCC', multiple='stack', height=15, aspect=25/15).set(title='A distribution plot of the soil resources in Nigeria and the states that have them', xlabel='States of Nigeria', ylabel='mg_code')


```

```python
url4 = "https://datasets.mojaglobal.workers.dev/0:/Bioclimatic&EcologicalZones/Global_Ecological_Zone_GEZ/byStateGEZ/NGA/NGA_ALL_states_GEZ.json"
f=requests.get(url4).content

data = gpd.read_file(io.StringIO(f.decode('utf-8')))
data
```

```python
df3 = data.groupby(['name', 'gez_name'])['gez_code'].sum().unstack().fillna(0)
df3
```

```python
fig, ax = plt.subplots(1, 1, figsize=(25, 15))
ax.bar(df3.index, df3['Tropical dry forest'], label='Tropical dry forest')
ax.bar(df3.index, df3['Tropical moist forest'], bottom=df3['Tropical dry forest'], label='Tropical moist forest')
ax.bar(df3.index, df3['Tropical mountain system'], bottom=df3['Tropical moist forest'], label='Tropical mountain system')
ax.bar(df3.index, df3['Tropical rainforest'], bottom=df3['Tropical mountain system'], label='Tropical rainforest')
ax.bar(df3.index, df3['Tropical shrubland'], bottom=df3['Tropical rainforest'], label='Tropical shrubland')
ax.bar(df3.index, df3['Water'], bottom=df3['Tropical shrubland'], label='Water')

plt.title('Graph showing the distribution of the Biological and Ecological Zones in Nigeria by states')
ax.legend()
```
